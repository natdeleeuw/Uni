Binomial Distribution
The binomial distribution models the number of successes in a fixed number of independent trials, each with the same probability of success. The probability mass function of a binomial distribution with parameters n (number of trials) and p (probability of success) is given by:

$P(X=k)=\binom{n}{k}p^{k}(1-p)^{n-k}$

<!-- The Bernoulli generator scheme can be used to generate random samples from a binomial distribution as follows:

Generate n independent Bernoulli random variables, each with probability of success p.
Count the number of successes among the n Bernoulli trials to get a random sample from the binomial distribution.
Here's a Jupyter Notebook implementation of the Bernoulli generator scheme for the binomial distribution: -->

the negative binomial distribution is a discrete probability distribution that models the number of failures in a sequence of independent and identically distributed Bernoulli trials with probability p of success, before a specified (non-random) number of successes (denoted 
r) occurs : 


$P(X=k)=\binom{k+r-1}{r-1}p^{r}(1-p)^{k}$




```python
# Question 1.1

def binomial_distribution(n, p, size):
    samples = np.zeros(size)
    for i in range(size):
        #does np.sum makes a difference?
        samples[i-1] = np.random.binomial(n, p)
    return samples
    
def negative_binomial_distribution(r, p, size):
    samples = np.zeros(size)
    for i in range(size):
        count = 0
        successes = 0
        while successes < r:
            count += 1
            bernoulli_sample = np.random.binomial(1, p)
            successes += bernoulli_sample
        samples[i -1] = count - successes
    return samples
    


samples = binomial_distribution(100, 0.3, 1000)

samples2 = binomial_distribution(50, 0.3, 1000)

samples3 = binomial_distribution(100, 0.5, 1000)

# samples4 = binomial_distribution(100, 0.3, 10000)


negative_samples = negative_binomial_distribution(100, 0.3, 1000)

negative_samples2 = negative_binomial_distribution(50, 0.3, 1000)

negative_samples3 = negative_binomial_distribution(100, 0.5, 1000)

# negative_samples4 = negative_binomial_distribution(100, 0.3, 10000)


plt.hist(samples)
plt.title("Binomial Distribution, n = 100, p = 0.3, size = 1000")
plt.show()

plt.hist(samples2)
plt.title("Binomial Distribution, n = 50, p = 0.3, size = 1000")
plt.show()

plt.hist(samples3)
plt.title("Binomial Distribution, n = 100, p = 0.5, size = 1000")
plt.show()

# plt.hist(samples4)
# plt.title("Binomial Distribution, n = 100, p = 0.3, size = 10000")
# plt.show()

print("______________")

plt.hist(negative_samples)
plt.title("Negative Binomial Distribution, r = 100, p = 0.3, size = 1000")
plt.show()

plt.hist(negative_samples2)
plt.title("Negative Binomial Distribution, r = 50, p = 0.3, size = 1000")
plt.show()

plt.hist(negative_samples3)
plt.title("Negative Binomial Distribution, r = 100, p = 0.5, size = 1000")
plt.show()

# plt.hist(negative_samples4)
# plt.title("Negative Binomial Distribution, r = 100, p = 0.3, size = 10000")
# plt.show()

```

```python
# Question 1.3
def coint_flip(n):
    
    samples = np.zeros(n)
    
    sum = 0
    for i in range(n):
        currentR = np.random.binomial(1, 0.5)
        if currentR > 0.5:
            sum += 1
        else:
            sum -= 1
        samples[i] = sum/np.sqrt(n)
    return samples

n = 1000
flips = coint_flip(n)
plt.plot(np.arange(0,n) /1000, flips)
plt.title("Sum of coins")
plt.show()

print("We see here that the total gain stays close to 0, this makes sense because there is as much chance to increment 1 then to decrement 1 ")
```
